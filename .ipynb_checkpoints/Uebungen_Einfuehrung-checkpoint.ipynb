### Datenstrukturen

```python
# Listen
# Empfehlenswert für gleichartige Werte

x = [1, 2, 3, 4, 5]
y = [555, 'world', 'helloWorld', 'xyz'] #nicht empfehlenswert, da int und string gemixed wird
print(x[0])
print(y[0])
print(y[1])

# Index kann negativ sein - gezählt wird dann vom Ende der Liste

print(x[-1])  # letztes Element
print(x[-5])  # 5. Element vom Ende gesehen - hier 1. Element
```

```python
# List Slicing (Rückgabe ist neue Liste)

# a[start:end] # von start bis end-1
# a[start:]    # von start bis zum Ende der Liste
# a[:end]      # vom Anfang der Liste bis end-1
# a[:]         # die gesamte Liste

# x = [1, 2, 3, 4, 5]
print(x[:2])
print(x[1:])
print(x[1:2])
print(x[1:-1])

# a[start:end:step]  # start und end wie oben, step ist Schrittweite

print(x[::2])
print(x[1::2])
print(x[1:2:2])
print(x[1:-1:2])
```

```python
# Listen sind mutable

x[1] = 10
x[0] = 0.0
print(x)
```

```python
# Möglich, aber nicht empfehlenswert

x = [1, 2, 'hallo', 0.0023456]
print(x)
```

### NumPy - *C Arrays in Python*

```python
import numpy as np
```

```python
#kleine Testfunktion:

def mySqrFunction(n):
    return n*n
    
```

```python
# Python-Liste und Numpy-Array mit jeweils 100 Millionen Int-Werten (0 - 99.999.999)

numbers = range(100000000)
np_numbers = np.arange(100000000)
```

```python
# List Comprehension - Pythonic
# %time zum Messen der Zeit

%time results = [mySqrFunction(n) for n in numbers]
print(results[:10])
```

```python
# Numpy - Funktional, vektorisiert und sehr effizient

%time results = np.apply_along_axis(mySqrFunction, 0, np_numbers)
print(results[:10])
```

```python
# Morphologische Operatoren in OpenCV: https://docs.opencv.org/3.4/db/df6/tutorial_erosion_dilatation.html
# Rekonstruktion mit Dilation in skimage: https://scikit-image.org/docs/dev/api/skimage.morphology.html#skimage.morphology.reconstruction

import scipy
from scipy.ndimage import gaussian_filter
from skimage import img_as_float
from skimage.morphology import reconstruction

# numpy Array (uint8) konvertieren in float (wichtig für die Substraktion im nächsten Schritt)
image = img_as_float(data.coins())
image = gaussian_filter(image, 1)

#Nachbarschaft/Maske definieren
seed = np.copy(image)
seed[1:-1, 1:-1] = image.min()
mask = image

#Morphologische Operation: Dilation
dilated = reconstruction(seed, mask, method='dilation')


#Ergebnisse plotten
fig, (ax0, ax1, ax2) = plt.subplots(nrows=1,ncols=3,figsize=(15,15))                                    

ax0.imshow(image)
ax0.set_title('original')
ax0.axis('off')

ax1.imshow(dilated, vmin=image.min(), vmax=image.max(), cmap='gray')
ax1.set_title('dilated')
ax1.axis('off')

ax2.imshow(image - dilated, cmap='gray')
ax2.set_title('image - dilated')
ax2.axis('off')
fig.tight_layout()

```

```python
#dieselbe Opration kann auch mittels OpenCV library durchgeführt werden
import cv2
from skimage import measure


kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
print(kernel)
res = cv2.morphologyEx((image-dilated),cv2.MORPH_OPEN,kernel)
#visualisierung
plt.figure(figsize=(15,8))
plt.imshow(res, cmap='gray')
plt.show()
```

### Interaktive Elemente (z.B. zur Festlegung von Funktions-Parametern)

```python
from io import BytesIO
from IPython.display import Image
from ipywidgets import interact, fixed
import matplotlib as mpl
from skimage import data, filters, io
```

```python
# Hilfsfunktion zur Konvertierung eines Numpy-Arrays in ein 'Image'-Objekt

def arr2img(arr):
    if arr.ndim == 2:
        format, cmap = 'png', mpl.cm.gray
    elif arr.ndim == 3:
        format, cmap = 'jpg', None
    else:
        raise ValueError("Only 2- or 3-d arrays can be displayed as images.")
    vmax = 255 if arr.dtype == 'uint8' else 1.0
    with BytesIO() as buffer:
        mpl.image.imsave(buffer, arr, format=format, cmap=cmap, vmin=0, vmax=vmax)
        out = buffer.getvalue()
    return Image(out)
```

```python
# Funktion zur Bildverarbeitung - Gaussfilter mit Farbkanalregler

def edit_image(image, sigma=0.1, R=1.0, G=1.0, B=1.0):
    new_image = filters.gaussian(image, sigma=sigma, multichannel=True)
    new_image[:,:,0] = R*new_image[:,:,0]
    new_image[:,:,1] = G*new_image[:,:,1]
    new_image[:,:,2] = B*new_image[:,:,2]
    return arr2img(new_image)
```

```python
# Bild zuweisen

img = data.astronaut()

# Aufruf der definierten Funktion

edit_image(img, sigma=1)
```

```python
# Aufruf mit anderen Parametern

edit_image(img, sigma=0, R=0.0)
```

```python
# Interaktiv wäre doch viel besser !!!

from IPython.display import display
import ipywidgets as widgets

# Definition von Paremeter-Grenzen und Schrittweite für die Farbkanäle.
# (untere Grenze, obere Grenze, Schrittweite)
limits = (0.0,1.0,0.01)

# Parameter-Grenzen, wobei die Änderung des Parameters nur
# bei mouse release events an die Funktion übergeben werden.
# Sinnvoll bei Funktionen mit langer Laufzeit! (hier nur als Beispiel)
sigma_limit = widgets.FloatSlider(min=0.0, max=10.0, step=0.01, continuous_update=False)

# Interaktives Widget initialisieren und ausführen
# image=fixed(img) - der Parameter image der Funktion edit_image soll nicht verändert werden
# 
interact(edit_image, image=fixed(img), sigma=sigma_limit, R=limits, G=limits, B=limits);
```
